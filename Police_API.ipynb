```python
# Fetch data in multiple requests with each request limited to 50000 results
# Take approx. 2 min to run each chunk (total 3 chunks)
# Ideally should use for loop to do these chucks but did not know how to dynamically name different df

police_json1 = requests.get("https://www.dallasopendata.com/resource/qv6i-rri7.json?$limit=50000&$offset=0").json()
police1 = json_normalize(police_json1) # Convert JSON to df

police_json2 = requests.get("https://www.dallasopendata.com/resource/qv6i-rri7.json?$limit=50000&$offset=50000").json()
police2 = json_normalize(police_json2)

police_json3 = requests.get("https://www.dallasopendata.com/resource/qv6i-rri7.json?$limit=50000&$offset=100000").json()
police3 = json_normalize(police_json3)

police_json4 = requests.get("https://www.dallasopendata.com/resource/qv6i-rri7.json?$limit=50000&$offset=150000").json()
police4 = json_normalize(police_json4)
```
